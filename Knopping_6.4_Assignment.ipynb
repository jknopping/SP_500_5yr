```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('SP500_5yr_data_weekly.csv')
```

```python
df = df.set_index('Date')
```

```python
df['% Change'] = ((df['Adj Close'] - df['Open']) / df['Open']) * 100
df
```

```python
plt.figure(figsize=(7, 5))
plt.plot(df['Close'], color = 'g')
plt.tick_params(axis ='both', which = 'both', length = 10) 
plt.tick_params(axis ='x', rotation = 90) 
plt.xticks(np.arange(0, 260, 12))
plt.title('S&P 500 five-year growth')
plt.ylabel('Currency in USD')
plt.show()
```

```python
'''
The figure above charts the growth of the S&P 500 over the past five years. 
The value of the stock has clearly risen over this time period, despite periods of volatility.
'''
```

```python
plt.figure(figsize=(6, 4))
plt.hist(df['% Change'], color = 'purple', bins = 20)
plt.tick_params(axis ='both', which = 'both', length = 5)
plt.yticks(np.arange(0, 50, 5))
plt.title('S&P 500 Weekly Growth')
plt.xlabel('% Change in Value')
plt.ylabel('Number of Weeks')
plt.show()
```

```python
'''
By showing the percentage change in value for the stock every week, the histogram above helps to visualize the 
volatility of the S&P 500 over the past five years. The value of the S&P 500 generally has a percent change 
between -2 and +2.5%, and has a net postitive percent change more often than not.
''' 
```

```python
df_five_ydt = df.iloc[0:52, 4:5]
df_four_ydt = df.iloc[52:104, 4:5]
df_three_ydt = df.iloc[104:156, 4:5]
df_two_ydt = df.iloc[156:208, 4:5]
df_one_ydt = df.iloc[208:260, 4:5]
#using legend()
```

```python
plt.figure(figsize=(10, 10))
plt.subplot(3, 2, 1)
plt.plot(df_five_ydt, color = 'r')
plt.tick_params(axis ='both', which = 'both', length = 10) 
plt.tick_params(axis ='x', rotation = 45) 
plt.xticks(np.arange(0, 52, 4))
plt.title('S&P 500 5 Years to Date')
plt.ylabel('Currency in USD')

plt.subplot(3, 2, 2)
plt.plot(df_four_ydt, color = 'b')
plt.tick_params(axis ='both', which = 'both', length = 10) 
plt.tick_params(axis ='x', rotation = 45) 
plt.xticks(np.arange(0, 52, 4))
plt.title('S&P 500 4 Years to Date')
plt.ylabel('Currency in USD')

plt.subplot(3, 2, 3)
plt.plot(df_three_ydt, color = 'g')
plt.tick_params(axis ='both', which = 'both', length = 10) 
plt.tick_params(axis ='x', rotation = 45) 
plt.xticks(np.arange(0, 52, 4))
plt.title('S&P 500 3 Years to Date')
plt.ylabel('Currency in USD')

plt.subplot(3, 2, 4)
plt.plot(df_two_ydt, color = 'purple')
plt.tick_params(axis ='both', which = 'both', length = 10) 
plt.tick_params(axis ='x', rotation = 45) 
plt.xticks(np.arange(0, 52, 4))
plt.title('S&P 500 2 Years to Date')
plt.ylabel('Currency in USD')

plt.subplot(3, 2, 5)
plt.plot(df_one_ydt, color = 'turquoise')
plt.tick_params(axis ='both', which = 'both', length = 10) 
plt.tick_params(axis ='x', rotation = 45) 
plt.xticks(np.arange(0, 52, 4))
plt.title('S&P 500 1 Year to Date')
plt.ylabel('Currency in USD')

plt.tight_layout()
plt.show()
```

```python
'''
The figure above includes five graphs, each representing the growth in the S&P 500 stock over the 
given year, as of 10/25/19. In each year, the value of the stock increased, while the period between
October 2016 and September 2017 represented the most stable year of the past five in terms of growth. 
In this set of visual data, there is a trending drop in the stock value around the last quarter and 
first quarter of the calendar year.
'''
```

```python
df_one_ydt = np.array(df.iloc[208:256, 6:7])
df_past_mo = np.array(df.iloc[256:260, 6:7])
plt.figure(figsize=(6,6))

plt.hist(df_one_ydt, density=True, color = 'blue', bins=np.arange(-8, 8), alpha = .5)
plt.hist(df_past_mo, density=True, color = 'red', bins=np.arange(-8, 8), alpha = .5)
plt.title('S&P 500 Year-to-Date vs. Past Month')
plt.xlabel('% Change in Value')

plt.show()
```

```python
'''
The histogram above shows the volatility of the S&P 500 during the past month (red), versus
the rest of the past year (blue). From this representaiton, it is clear that this past month
has experienced less volatility than the rest of year, so far.
'''
```
